```python
def print_sample(size=100):
    samples = torch.tensor([int(x) for x in data['anns'].keys()])[torch.randint(high=len(data['anns']), size=(size,))]
    for i in samples.numpy():
        try:
            img = Image.open(f'../datasets/coco_text_imgs/{i}.jpg')
            boxes = open(f'../datasets/coco_text_imgs/{i}.txt').read()
        except: continue
        w, h = img.size
        img1 = ImageDraw.Draw(img)  
        for i in boxes.split('\n'):
            _, x, y, x_h, y_h = [float(x) for x in i.split(' ')]
            mid = (w*x, h*y)
            size = (w*x_h/2, h*y_h/2)
            shape = [(mid[0]-size[0], mid[1]-size[1]), (mid[0]+size[0], mid[1]+size[1])]
            img1.rectangle(shape, outline ="red")
        img
        plt.imshow(img)
        plt.show()

print_sample(size=100)
```
